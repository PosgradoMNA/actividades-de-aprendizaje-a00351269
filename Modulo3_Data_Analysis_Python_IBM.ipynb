#5. Correlacion y Causación


Correlación: es una medida de interdependencia entre variables.

Causacion: es la relación entre causa y efecto entre dos variables.


Es importante saber la diferencia.
Correlacion no implica Causacion.

Determinar la correlacion es mas simple que determinar la causacion. ya que esta ultima requiere de experimentacion.




**Correlación de Pearson**

La Correlacion de Pearson mide la dependencia lineal entre dos variables X y Y.

El coeficiente resultante es un valor entre -1 y 1 inclusive, donde:

1: es una correlacion positiva perfecta.

0: No HAY correlacion lineal , las dos variable no se afecta la una a la otra.

-1: es una correlacion negativa perfecta.



```python
df.corr()# la correlacion de pearson es el valor por efecto del metodo .corr()
```

P-valor (P-value)

Que es  P-value? 

El P-value es el valor de la probabilidad de que la correlacion entre dos variables sea estadisticamente significativo.

Normalmente , escogemos un nivel de significancia de 0.05, lo que quieres decir que con el 95% de confianza, la correlacion entre las variables es significativa.

por convencion:


si el **p-value < 0.001**:  decimos que hay **fuerte** evidencia de que la **correlacion** es significativa.


si el **p-value  < 0.05**: 
  existe una **moderada** evidencia de que la **correlacion** es significativa.


si el **p-value  < 0.1**: es evidencia de una **débil** evidencia de que la **correlacion** es significativa.

si el **p-value  > 0.1**: **No hay** evidencia de que la **correlacion** es significativa.

```python
from scipy import stats
```

#6. ANOVA

ANOVA: Analisis de Varianza

(ANOVA) es un metodo estadistico utilizado para probar si hay diferencias significatuvas entre dos o mas grupos.

ANOVA regresa 2 parámetros:

**F-test score: **
ANOVA asume que la media de todos los grupos es la misma, calcula cuanto la media actual se desvia del supuesto y reporta esto como un F-test score. 
Un F-test  score grande significa que hay gran diferencia entre las medias.


**P-value**: P-value dice que tan estadisticamente significativo es el score calculado .

SI nuestra variable price variable está fuertemente correlacionada con la variable que analizaremos , esperamos ANOVA reglres un F-test score medible y un valor pequeño de P-value.

```python
grouped_test2=df_gptest[['drive-wheels', 'price']].groupby(['drive-wheels'])
grouped_test2.head(2)
```
