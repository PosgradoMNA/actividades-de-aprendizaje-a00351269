#Data Acquisition

Existen varios formatos de un dataset:  .csv, .json, .xlsx , etcétera. 
El dataset puede estar almacenado en diferentes lugares, como una maquina local o en internet o en la nube.

Vamos a ver el dataset ejemplo llamado "Automobile Dataset" el cual es una fuente de archivo en formato CSV (comma separated value), el cual utilizaremos para la practica de lectura.

Data source: https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data.csv
Data type: csv


```python
import pandas as pd
import numpy as np
```

```python
path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
df = pd.read_csv(path, header=None)
df.head()
```

```python
# Write your code below and press Shift+Enter to execute 
df.tail(10)
```

```python
# Creando los Headers (encabezados) a partir de una lista , se pueden utilizar comillas simples ' o dobles "
headers = ['symboling','normalized-losses','make','fuel-type','aspiration', 'num-of-doors','body-style',
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
df.columns = headers #asignando los nombres de la lista de headers a las columnas del dataframe
df.head(20)
```

```python
df2=df[df['normalized-losses']== "?"] #para ver que registros tenemos con "?"
df2
```

Tenemos que reemplazar el simbolo "?" con NaN para que el metodo dropna() puede remover los valores perdidos

```python
df1=df.replace('?',np.NaN)
```

```python
df1.describe(include="all") #algunos estadisticos para ver que algunas columnas tienen datos nulos.
```

```python
df=df1.dropna(subset=["price"], axis=0) #borramos los reglones con precio en nulo
df.head(20)
```

```python
df.columns #los nombres de las columnas del dataset son los indices de las columnas en nombre, empiezan de 0 por numero
```

#Guardando el Dataset

Pandas nos permite guardar el Dataset a csv. 

Utilizando el método **dataframe.to_csv()** podemos guardar el archivo con el **path del archivo** y el nombre entre comillas **"nombre.csv"**.

por ejemplo con el **dataframe df  como  automobile.csv**  a la maquina local , dende **index = False** significa que los nombre de cada columna no se escribiran.

```python
df.to_csv("automobile.csv", index=False)
```
